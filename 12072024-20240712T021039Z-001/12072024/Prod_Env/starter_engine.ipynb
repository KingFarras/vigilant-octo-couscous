```python
# prompt: gimme a button in streamlit

import streamlit as st

# Create a button widget.
button = st.button("Click me!")

# Check if the button is clicked.
if button:
    # Do something when the button is clicked.
    st.write("You clicked the button!")
```

```python
# prompt: ok now. add more button with some logic

# Create two buttons.
button1 = st.button("Button 1")
button2 = st.button("Button 2")

# Add logic to the buttons.
if button1:
    # Do something when button 1 is clicked.
    st.write("You clicked button 1!")
elif button2:
    # Do something when button 2 is clicked.
    st.write("You clicked button 2!")

# Add a third button that only appears if both other buttons have been clicked.
if button1 and button2:
    button3 = st.button("Button 3")
    if button3:
        # Do something when button 3 is clicked.
        st.write("You clicked button 3!")
```
